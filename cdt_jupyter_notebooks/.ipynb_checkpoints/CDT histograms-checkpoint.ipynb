```python
import matplotlib.pyplot as plt
import pandas as pd
plt.close("all")

data_flat = pd.read_csv('Downloads/Volume_[32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, \
32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32, \
32, 32]_statistical (1).csv')

flat_actions = data_flat.iloc[:, 1]

print(flat_actions)

def plot_histogram(data, bins=1000, figsize=(16, 12)):
    plt.figure(figsize=figsize)
    plt.hist(data, bins=bins, color='grey')
    plt.xlabel('Action')
    plt.ylabel('Frequency')
    plt.show()
    
plot_histogram(flat_actions)

```
