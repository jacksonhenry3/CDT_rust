```python
import matplotlib.pyplot as plt
import pandas as pd
plt.close("all")

data_flat = pd.read_csv('../data/Statistical_Volume_8192_TimeSize_64_stepsize_2001_sample_2.csv')

flat_actions = data_flat.iloc[:, 1]

print(flat_actions)

def plot_histogram(data, bins=1000, figsize=(16, 12)):
    plt.figure(figsize=figsize)
    plt.hist(data, bins=bins, color='grey')
    plt.xlabel('Action')
    plt.ylabel('Frequency')
    plt.show()
    
plot_histogram(flat_actions)

```
